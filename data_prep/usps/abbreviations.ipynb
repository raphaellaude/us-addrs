## Street Abbreviations (C1)


```python
st_abbrevs = pd.read_html(os.path.join(PATH_TMP, 'C1 Street Suffix Abbreviations _ Postal Explorer.html'), skiprows=3)[1]
st_abbrevs.columns = ['primary_name', 'common_abbrevs', 'standard_abbrev']
st_abbrevs.dropna(inplace=True)
```

```python
# regex remove all non capital letter chars
cleaned_abbrevs = st_abbrevs['standard_abbrev'].str.replace(r'[^A-Z]', '', regex=True)
```

```python
st_abbrevs[cleaned_abbrevs.ne(st_abbrevs['standard_abbrev'])]
```

```python
cleaned_abbrevs[st_abbrevs['primary_name'].eq('BAYOU')]
```

```python
st_abbrevs['standard_abbrev'] = cleaned_abbrevs
```

```python
st_abbrevs.head()
```

```python
st_abbrevs[~st_abbrevs['primary_name'].isin(st_abbrevs['common_abbrevs'].values)]
```

```python
st_abbrevs_to_print = pd.concat([
    st_abbrevs[['primary_name', 'standard_abbrev']].rename(
        columns={"primary_name": "common_abbrevs"}),
    st_abbrevs[['common_abbrevs', 'standard_abbrev']]
    ]).drop_duplicates()
```

```python
for ix, r in st_abbrevs_to_print.iterrows():
    print(f'\t\tm.insert("{r[0]}", "{r[1]}");')
```
